```python
# data from https://allisonhorst.github.io/palmerpenguins/

import matplotlib.pyplot as plt
import numpy as np

Models = ("RECOVER", "Scale mixture", "Laplace", "single Gaussian", "Dropout")
model_times = {
    'UCB Acquisition': (23040 , 2220, 5040, 4200, 4500),
    'Greedy Acquisition': (23400, 2313, 4680, 5354, 4714 ),
    'Random Acquisition': (15600, 2440, 3807, 3961 ,5145)
}

x = np.arange(len(Models))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(figsize=(15, 6))#(layout='constrained')

for attribute, measurement in model_times.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3, rotation='vertical')
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Time (s)')
ax.set_xlabel('Models')
ax.set_title('SMO compute time for models')
ax.set_xticks(x + width, Models, rotation='vertical')
ax.legend()
ax.set_ylim(0, 28000)


plt.show()
```
